```python
import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine

import getpass
```

```python
#Question 1
query = 'SELECT COUNT(DISTINCT last_name) AS "Nº different last names"\
FROM sakila.actor;'
data = pd.read_sql_query(query, engine)
data.head()
```

```python
#Question 2
query = 'SELECT *, DATE_FORMAT(rental_date, "%%b") AS "month", \
	DATE_FORMAT(rental_date, "%%a") AS "weekday", \
CASE \
WHEN DATE_FORMAT(rental_date, "%%a") = "Sat" then "weekend" \
WHEN DATE_FORMAT(rental_date, "%%a") = "Sun" then "weekend" \
ELSE "workday" \
END AS "day_type" \
FROM sakila.rental;'
data = pd.read_sql_query(query, engine)
data.head()
```

```python
#Question 3
query = 'SELECT title \
FROM sakila.film \
WHERE title LIKE "ARMAGEDDON%%";'
data = pd.read_sql_query(query, engine)
data.head()
```

```python
#Question 4
query = 'SELECT title, length FROM sakila.film \
ORDER BY length DESC \
LIMIT 10;'
data = pd.read_sql_query(query, engine)
data.head(10)
```

```python
#Question 5
query = 'SELECT COUNT(special_features) AS "Nº films"\
FROM sakila.film \
WHERE special_features LIKE "%%Behind the Scenes%%";'
data = pd.read_sql_query(query, engine)
data.head()
```

```python
#Question 6
query = 'SELECT rating, ROUND(AVG(length),2) AS "Mean_length" FROM sakila.film \
GROUP BY rating \
HAVING AVG(length) > 120;'
data = pd.read_sql_query(query, engine)
data.head()
```

```python
#Question 7
query = 'SELECT title, length, \
DENSE_RANK() OVER(ORDER BY length ASC) AS "Rank" \
FROM sakila.film \
WHERE length IS NOT NULL AND (length > 0);'
data = pd.read_sql_query(query, engine)
data.head(60)
```
